<h1>Built-in functions</h1>

Python has a set of built‑in functions that are always available without needing to import any module. Examples include print(), len(), sum(), abs(), sorted(), and many more.

<h2> Math & Numbers</h2>

```python
x =-12 
abs(x)
```

```python
x = 2
pow(x,2)
```

```python
round(3.14159, 2)
```

```python
divmod(17, 5)
```

```python
min(4, 7, 2)     # → 2
max([1, 9, 3])   # → 9
sum([1, 2, 3])   # → 6
```

<h2>Data Types & Conversion</h2>


```python
int("42")        # → 42
float("3.14")    # → 3.14
complex(2, 3)    # → (2+3j)
str(123)         # → '123'
bool(0)          # → False
list("abc")      # → ['a', 'b', 'c']
dict(a=1, b=2)   # → {'a': 1, 'b': 2}
tuple([1, 2, 3]) # → (1, 2, 3)
set([1, 2, 2])   # → {1, 2}
bin(10)          # → '0b1010'
oct(10)          # → '0o12'
hex(255)         # → '0xff'
ord("A")         # → 65
chr(65)          # → 'A'
```

<h2>Object Inspection</h2>


```python
type("hello")        # → <class 'str'>
isinstance(5, int)   # → True
dir([])              # → ['append', 'clear', 'copy', ...]
id(42)               # → unique memory address (e.g., 9794400)\
callable(len)        # → True
```

<h2>Iteration & Functional Tools</h2>

```python
range(5)             # → range(0, 5) → [0,1,2,3,4]
list(enumerate("abc")) # → [(0, 'a'), (1, 'b'), (2, 'c')]
list(zip([1,2], ['a','b'])) # → [(1, 'a'), (2, 'b')]
list(map(str.upper, ["a", "b"])) # → ['A', 'B']
list(filter(lambda x: x>2, [1,2,3,4])) # → [3,4]
sorted([3,1,2])      # → [1,2,3]
```

<h2>Input/Output</h2>


```python
print("Hello World") # → Hello World
# input() would wait for user input, e.g.:
# name = input("Enter your name: ")
# open a file:
f = open("test.txt", "w")
f.write("Hello File")
f.close()
```

```python
eval("2 + 3")        # → 5
exec("x = 10")       # defines variable x
globals().keys()     # → shows global variables
locals()             # → shows local variables
__import__("math").sqrt(16) # → 4.0
```
